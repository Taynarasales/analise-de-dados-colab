```python
import requests
import pandas as pd


def buscar_clima(cidade, latitude, longitude):
    url = "https://api.open-meteo.com/v1/forecast"
    dias = 20

    # Parâmetros da requisição
    parametros = {
        "latitude": latitude,
        "longitude": longitude,
        "daily": "temperature_2m_max,precipitation_sum",
        "past_days": dias,
        "timezone": "auto"
    }

    # Requisição à API
    resposta = requests.get(url, params=parametros)

    if resposta.status_code == 200:
        dados = resposta.json()
        dados_diarios = dados["daily"]

        # Criando o DataFrame com os nomes exigidos
        df = pd.DataFrame({
            "time": dados_diarios["time"],
            "temperature_2m_max": dados_diarios["temperature_2m_max"],
            "precipitation_sum": dados_diarios["precipitation_sum"]
        })

        # Convertendo para Fahrenheit
        df["temp_fahrenheit"] = (df["temperature_2m_max"] * 9/5) + 32

        # Exibindo os resultados
        print(f"\nDados climáticos de {cidade} (últimos {dias} dias):\n")
        print(df.to_string(index=False))

    else:
        print("Erro ao acessar a API.")
        print("Código do erro:", resposta.status_code)
        print(resposta.text)


# Entrada de dados do usuário
cidade = input("Digite sua cidade: ")
latitude = float(input("Digite a latitude: "))
longitude = float(input("Digite a longitude: "))

buscar_clima(cidade, latitude, longitude)
```
